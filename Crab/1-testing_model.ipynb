```python
import numpy as np
import astropy.units as u
import matplotlib.pyplot as plt
import yaml

from gammapy.datasets import MapDataset, Datasets, FluxPointsDataset
from gammapy.maps import Map
from gammapy.makers import FoVBackgroundMaker
from gammapy.catalog import SourceCatalogHGPS, SourceCatalogGammaCat, SourceCatalog4FGL 
from gammapy.modeling.models import Models, SkyModel, TemplateSpatialModel, PowerLawNormSpectralModel
from gammapy.modeling import Fit
from gammapy.estimators import ExcessMapEstimator, FluxPoints, FluxPointsEstimator
```

```python
hesseras=['hess1', 'hess2', 'hess1u']
muoneff_flag= True
edisp = True

info_list = [] #hessera, muoneff, edisp, name

for hessera in hesseras:
    if hessera =='hess1u':
        muoneff_flag=False
        
    if hessera == 'hess1':
        for within_idx in range(3):
            name = f'20220711_{hessera}_muoneff{muoneff_flag}_edisp{edisp}_{within_idx}'
            info_list.append([hessera, muoneff_flag, edisp, name])
    else:
        name = f'20220711_{hessera}_muoneff{muoneff_flag}_edisp{edisp}'
        info_list.append([hessera, muoneff_flag, edisp, name])        
```

```python
from gammapy.modeling.models import GaussianSpatialModel, LogParabolaSpectralModel, PointSpatialModel, create_crab_spectral_model


models_list = []
dataset_list = []
for idx in range(len(info_list)):
    print(idx, info_list[idx][3])
    dataset= MapDataset.read(f'{info_list[idx][3]}.fits')   
    
    spatial=PointSpatialModel(lon_0=83.63*u.deg, lat_0=22.01*u.deg, frame='icrs')
    crab_source= SkyModel(spatial_model=spatial, spectral_model=create_crab_spectral_model('hess_ecpl'), name='Crab')

    dataset.models = Models([crab_source])
    dataset = FoVBackgroundMaker().make_default_fov_background_model(dataset=dataset)
    
    dataset.npred_signal().sum_over_axes().plot(add_cbar=True)
    plt.show()
    
    result0= Fit().run(datasets=[dataset])
    print(result0)
    print(dataset.models)
    models_list.append(dataset.models.copy())
    dataset_list.append(dataset)
    print(f'finished fit {idx}')
```

```python
colors = ['navy', 'C0', 'C1', 'C2', 'C5']

x_axis = dataset.counts.geom.axes['energy'].center.value
energy_bounds = [0.3, 100] * u.TeV
sed_type = "e2dnde"
```

```python

fig, ax = plt.subplots(1,1, figsize=(4, 3), dpi=100)

for idx, models in enumerate(models_list):
    models[0].spectral_model.plot(ax=ax, color=colors[idx],
        energy_bounds=energy_bounds, sed_type=sed_type, label=f'dataset-{idx}')
    models[0].spectral_model.plot_error(ax=ax, energy_bounds=energy_bounds, sed_type=sed_type)
plt.legend(loc='lower left')
plt.ylim(1e-13, 2e-10)
plt.title('Crab')
```

```python
fig, ax = plt.subplots(1,4, figsize=(20, 3), dpi=100)

for idx in range(len(models_list)-1):
    create_crab_spectral_model('hess_ecpl').plot(ax=ax[idx], color='black',
        energy_bounds=energy_bounds, sed_type=sed_type, label=f'catalog')
    
    models_list[idx][0].spectral_model.plot(ax=ax[idx],
        energy_bounds=energy_bounds, sed_type=sed_type, label=f'present time bin')
    models_list[idx][0].spectral_model.plot_error(ax=ax[idx], energy_bounds=energy_bounds, sed_type=sed_type)

    models_list[idx+1][0].spectral_model.plot(ax=ax[idx],color=colors[idx+1],
        energy_bounds=energy_bounds, sed_type=sed_type, label=f'next time bin', ls='--')
    models_list[idx+1][0].spectral_model.plot_error(ax=ax[idx], energy_bounds=energy_bounds, sed_type=sed_type)
    ax[idx].set_ylim(1e-13, 2e-10)
    ax[idx].legend(loc='lower left')
plt.suptitle('Crab')
```

```python
for idx, model in enumerate(models_list):
    with open(f'crab_fitted_model_{idx}.yml', 'w') as outfile:
        yaml.dump(model.to_dict(), outfile, default_flow_style=False)
```

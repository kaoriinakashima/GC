```python
import numpy as np
import astropy.units as u
import matplotlib.pyplot as plt
import yaml

from gammapy.estimators import FluxPoints
```

```python
hesseras=['hess1', 'hess2', 'hess1u']
muoneff_flag= True
edisp = True

fp_list= []
for hessera in hesseras:
    if hessera =='hess1u':
        muoneff_flag=False

    fp = FluxPoints.read('fp_'+f'20220702_{hessera}_muoneff{muoneff_flag}_edisp{edisp}.fits')
    fp_list.append(fp)
```

```python
sed_type = "e2dnde"

for i, hessera in enumerate(hesseras):
    fp_list[i].plot(sed_type=sed_type, label=hessera)
    
plt.legend()
plt.ylim(1e-15, 1e-11)
```
